```python
import pandas as pd

# Load spreadsheet
accuracy = pd.read_csv('summary_metrics.csv', delimiter=';')

```

```python
accuracy.head()
```

```python
print(accuracy.dtypes)
```

```python
accuracy[['PCC', 'MSE', 'MARE', '95th percentile', 'MAE']] = accuracy[['PCC', 'MSE', 'MARE', '95th percentile', 'MAE']].replace(',', '.', regex=True).astype('float64')
```

```python
accuracy.head()
```

```python
import matplotlib.pyplot as plt
import numpy as np



FNN = accuracy.loc[0, 'PCC']
CNN = accuracy.loc[1, 'PCC']
RNN = accuracy.loc[2, 'PCC']
COMB = accuracy.loc[3, 'PCC']
CNN_mod = accuracy.loc[4, 'PCC']
RNN_mod = accuracy.loc[5, 'PCC']
COMB_mod = accuracy.loc[6, 'PCC']




# Define the labels and heights for the bars
labels = ['FNN', 'CNN', 'RNN', 'COMB', 'CNN_mod', 'RNN_mod', 'COMB_mod']
heights = [FNN, CNN, RNN, COMB, CNN_mod, RNN_mod, COMB_mod]

# Create an array with the positions of the bars on the x-axis
x_pos = np.arange(len(heights))

# Find the index of the highest value
max_index = np.argmax(heights)

# Create a list of edge colors, with a different color for the highest value
edgecolors = ['none' if i != max_index else 'indigo' for i in range(len(heights))]

# Create a list of line widths, with a different width for the highest value
linewidths = [0 if i != max_index else 2 for i in range(len(heights))]
# Create a bar chart
plt.bar(x_pos, heights, color=['teal', 'aquamarine', 'lightblue', 'lavender', 'pink', 'plum', 'purple'], edgecolor=edgecolors, linewidth=linewidths)

# Replace the x ticks with the labels
plt.xticks(x_pos, labels, rotation=45)

# Set the limits of the y-axis
plt.ylim([0.980, 1])

# Add a title and labels
plt.title('PCC')
plt.ylabel('PCC')

# Display the bar chart
plt.show()
```
